# Markov Chain

```python
import requests
from bs4 import BeautifulSoup
import random
from shmarkov import *
```

### Get text from Yahoo answer

```python
def getRandomQuestion():
    html = requests.get("https://answers.yahoo.com/answer").text
    soup = BeautifulSoup(html, 'html.parser')
    questions = soup.select(".qHead a")

    question = random.choice(questions).text
    
    return question
```

### Put all questions in one string

```python
text = ""
for i in range(30):
    text += getRandomQuestion() + " "
words = text.split()
```

```python
text
```

#### Make a Markov model in characters

##### ngram = 2

```python
model = markov_model(2, text)
```

```python
model
```

```python
for i in range(10):
    output_text = "".join(gen_from_model(2, model))
    print("test " + str(i) +": " +output_text)
```

##### ngram = 5

```python
model = markov_model(5, text)
```

```python
for i in range(10):
    output_text = "".join(gen_from_model(5, model))
    print("test " + str(i) +": " +output_text)
```

#### Make a Markov model in words

##### ngram = 2

```python
word_model = markov_model(2, words)
```

```python
word_model
```

```python
for i in range(10):
    output_text = " ".join(gen_from_model(2, word_model))
    print("test "+ str(i)+": " +output_text)
```

##### ngram = 5

```python
word_model = markov_model(5, words)
```

```python
for i in range(10):
    output_text = " ".join(gen_from_model(5, word_model))
    print("test "+ str(i)+": " +output_text)
```

### Put questions in lists

```python
questions = []
for i in range(30):
    questions.append(getRandomQuestion())
```

##### ngram = 2

```python
new_word_model = markov_generate_from_sequences(2, questions, 10)
```

```python
for item in new_word_model:
    print(''.join(item))
```

```python
new_word_model = markov_generate_from_sequences(5, questions, 10)
```
